```python
from descent import backward_prop,cost_comp,forward_prop,predict,update_params
```

```python
from inits import initialize_params_he,initialize_params_random,initialize_params_zero
```

```python
def model(X,Y,initialization,learning_rate = 0.01,iters = 10000):
    layer_size = [X.shape[0],10,5,1]
    cost_cache = []
    if initialization == "zero":
        params = initialize_params_zero(layer_size)
    if initialization == "random":
        params = initialize_params_random(layer_size)
    if initialization == "he":
        params = initialize_params_he(layer_size)
        
    for i in range(0,iters):
        A3,cache = forward_prop(X,params)
        cost = cost_comp(A3,Y)
        grads = backward_prop(X,Y,cache)
        params = update_params(params,grads,learning_rate)        
        if i%1000 == 0:
            print("Cost after iter",i,":",cost)
            cost_cache.append(cost)
    plt.plot(cost_cache)
    return params
```

```python
train_x,train_y,test_x,test_y = load_dataset()
```

```python
params = model(train_x,train_y,initialization = "zero")
print("train set:")
trainpred = predict(train_x,train_y,params)
print("test-set")
testpred = predict(test_x,test_y,params)
```

```python
params = model(train_x,train_y,initialization="random")
print("train set:")
trainpred = predict(train_x,train_y,params)
print("test-set")
testpred = predict(test_x,test_y,params)
```

```python
params = model(train_x,train_y,initialization="he")
print("train set:")
trainpred = predict(train_x,train_y,params)
print("test-set")
testpred = predict(test_x,test_y,params)
```
